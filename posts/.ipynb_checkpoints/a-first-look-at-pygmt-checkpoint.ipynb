# A first look at pygmt

```python
%matplotlib inline
from matplotlib import pyplot as plt 
```

```python
import xarray as xr 
from cartopy import crs as ccrs
```

```python
import pygmt
```

```python
grid = pygmt.datasets.load_earth_relief(
    resolution="02m", region=[160, 182, -50, -30], registration="gridline"
)
grid
```

```python
f, ax = plt.subplots(subplot_kw=dict(projection=ccrs.PlateCarree(central_longitude=180)))
grid.plot(ax=ax, transform=ccrs.PlateCarree())
ax.coastlines('10m')
```
